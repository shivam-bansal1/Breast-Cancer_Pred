```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.set_option('max_columns',None)
import warnings 
warnings.filterwarnings('ignore' )
```

```python
train_df = pd.read_csv('train.csv')
train_df.head()
```

```python
train_df.info()
```

```python
train_df.drop(['Unnamed: 32','id'],inplace=True,axis=1)
```

```python
train_df.duplicated().sum()
```

```python
train_df.describe().T
```

```python
#Checking for null values

plt.figure(figsize=(7,4))
sns.heatmap(train_df.isnull(),yticklabels=False)
```

```python
train_df.nunique()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
sns.countplot(train_df['diagnosis'],palette='magma')
plt.subplot(1,2,2)
train_df['diagnosis'].value_counts().plot.pie(autopct='%1.1f%%',radius=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_df['diagnosis']= le.fit_transform(train_df['diagnosis'])
```

```python
sns.pairplot(train_df.iloc[:,:11],hue='diagnosis')
```

```python
#Plotting to observe correlation between variables

plt.figure(figsize=(10,10))
sns.heatmap(train_df.iloc[ :,0:11].corr(),annot=True,fmt='.0%')
```

```python
#Scaling the values 

from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
scale.fit_transform(train_df)
```

```python
#Splitting the data

X = train_df.drop(['diagnosis'],axis=1)
Y = train_df.diagnosis

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=5)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train,Y_train)
print(f"Score :{log.score(X_train,Y_train)}")
pred = log.predict(X_test)

from sklearn.metrics import classification_report,accuracy_score
print(f'classification report{classification_report(Y_test,pred)}')
accuracy_score(Y_test,pred)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
error_train= []
error_test= []

for k in range(1,15):
    knn= KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,Y_train)
    y_pred_train = knn.predict(X_train)
    error_train.append(np.mean(Y_train!= y_pred_train))
    y_pred_test = knn.predict(X_test)
    error_test.append(np.mean(Y_test!= y_pred_test))
    
# plt.figure(figsize(10,5))
plt.plot(range(1,15),error_train,label="train")
plt.plot(range(1,15),error_test,label="test")
plt.xlabel('k Value')
plt.ylabel('Error')
plt.legend()
```

```python
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train,Y_train)
print(f"Score :{knn.score(X_train,Y_train)}")
pred = knn.predict(X_test)

print(f'classification report{classification_report(Y_test,pred)}')
accuracy_score(Y_test,pred)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,Y_train)
print(f"Score :{nb.score(X_train,Y_train)}")
pred = nb.predict(X_test)

print(f'classification report{classification_report(Y_test,pred)}')
accuracy_score(Y_test,pred)
```

### SVM

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(X_train,Y_train)
print(f"Score :{sv.score(X_train,Y_train)}")
pred = sv.predict(X_test)

print(f'classification report{classification_report(Y_test,pred)}')
accuracy_score(Y_test,pred)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train,Y_train)
print(f"Score :{tree.score(X_train,Y_train)}")
pred = tree.predict(X_test)

print(f'classification report{classification_report(Y_test,pred)}')
accuracy_score(Y_test,pred)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
Forest = RandomForestClassifier(criterion='entropy',random_state=0)
Forest.fit(X_train,Y_train)
print(f"Score :{Forest.score(X_train,Y_train)}")
pred = Forest.predict(X_test)

print(f'classification report{classification_report(Y_test,pred)}')
accuracy_score(Y_test,pred)
```

### XGBoost

```python
from xgboost import XGBClassifier
boost = XGBClassifier(random_state=0)
boost.fit(X_train,Y_train)
print(f"Score :{boost.score(X_train,Y_train)}")
pred = boost.predict(X_test)

print(f'classification report{classification_report(Y_test,pred)}')
accuracy_score(Y_test,pred)
```

```python
# Cross validation

from sklearn.model_selection import cross_val_score

cross_validation = cross_val_score(estimator = boost, X = X_train, y = Y_train, cv = 10)
print("Cross validation of XGBoost model = ",cross_validation)
print("Cross validation of XGBoost model (in mean) = ",cross_validation.mean())
```

```python
import pickle
 
# save model
pickle.dump(boost, open('breast_cancer_detector.pickle', 'wb'))
 
# load model
breast_cancer_detector_model = pickle.load(open('breast_cancer_detector.pickle', 'rb'))
 
# predict the output
y_pred = breast_cancer_detector_model.predict(X_test)
 
# Classification Report
print('Classification Report of XGBoost model: \n',classification_report(Y_test, y_pred),'\n')
 
# show the accuracy
print('Accuracy of XGBoost model = ',accuracy_score(Y_test, y_pred))
```
